# Predator Prey Model

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate,optimize
import sympy
```

```python
N1 = 10
N2 = 10
alpha = 2
beta = 1
gamma = 1
delta = 2
k=1
```

### Calculating roots


```python
def calc(x):
    f = [alpha*x[0] - beta*x[0]*x[1],  gamma*x[0]*x[1] - delta*x[1]]
    return f
for i in range(N1):
    print('\nN1:',i,', N2:',i,'\nRoots:',optimize.fsolve(calc, [i,i+1]))
```

Evolution

```python
t=np.arange(1,20, 0.1)

#starting of [N1,N2]
x=[5,2]
evolve = integrate.odeint(lambda x,t: [alpha*x[0] - beta*x[0]*x[1],  gamma*x[0]*x[1] - delta*x[1]],  x, t)

plt.plot(evolve[:, 0])
plt.title("N1 vs time")
plt.xlabel("Time")
plt.ylabel("Population of N1")
plt.show()
plt.plot(evolve[:, 1])
plt.title("N2 vs time")
plt.xlabel("Time")
plt.ylabel("Population of N2")
plt.show()
plt.plot(evolve[:, 0],label="N1")
plt.plot(evolve[:, 1],label="N2")
plt.title("N1 and N2 vs time")
plt.xlabel("Time")
plt.ylabel("Population of N1 and N2")
plt.legend()
plt.show()


for i in range(4):
    x=[i+4,i+2]
    evolve = integrate.odeint(lambda x,t: [alpha*x[0] - beta*x[0]*x[1],  gamma*x[0]*x[1] - delta*x[1]],  x, t)

    #Phase-space plot
    plt.plot(evolve[:,0],evolve[:, 1],label="N1: "+str(x[0])+", N2:"+str(x[1]))
    
plt.title("Phase space plot of N1,N2")
plt.xlabel("N1")
plt.ylabel("N2")
plt.legend()
plt.show()
```

Trajectory

```python
for x in np.arange(0, N1, 0.6):
    for y in np.arange(0, N2, 0.6):
        r=[alpha*x - beta*x*y, gamma*x*y - delta*y]
        plt.arrow(x, y, r[0]/N1, r[1]/N2)
plt.xlim(0,N1)
plt.title("Trajectory")
plt.xlabel("N1")
plt.ylabel("N2")
plt.ylim(0,N2)
plt.show()  
```

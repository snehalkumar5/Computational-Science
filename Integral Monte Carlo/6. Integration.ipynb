# Integration using Monte Carlo

```python
from matplotlib import pyplot as plt
import numpy as np
import random

def I_pnt(n):
    print('N:',n)
    y = []
    for N in range(1, n+1):
        dx = 1/N
        I1 = sum([3*(random.random()**2)*dx for __ in range(N)])
        I2 = 0
        y.append(I1)
    plt.plot(y,label="Simulated")
    plt.plot([1 for i in range(n)],label="Actual")
    plt.title('I vs Number of Points')
    plt.xlabel("N")
    plt.ylabel("I")
    plt.legend()
    plt.show()
    
def I_pnt2(n):
    print('N:',n)
    y = []
    for N in range(1, n+1):
        dx = 1/N
        I1 = sum([3*(random.random()**2)*dx for __ in range(N)])
        I2 = 0
        for __ in range(N):
            yy = random.random()
            i = sum([(random.random()**2)*yy*dx for ___ in range(N)])
            I2 += i * dx
        y.append(I2)
    plt.plot(y,label="Simulated")
    plt.plot([1/6 for i in range(n)],label="Actual")
    plt.title('I vs Number of Points')
    plt.xlabel("N")
    plt.ylabel("I")
    plt.legend()
    plt.show()
    
def I_trials(N,trials):
    print('N:',N,' Trials:',trials)
    y = []
    for _ in range(1,trials+1):
        dx = 1/N
        I1 = sum([3*(random.random()**2)*dx for __ in range(N)])
        I2 = 0
        y.append(I1)
    print('Integral:',y[-1])
    print('Standard deviation:',np.std(y))
    plt.plot(y,label="Simulated")
    plt.plot([1 for i in range(trials)],label="Actual")
    plt.xlabel("Trials")
    plt.ylabel("I")
    plt.legend()
    plt.title('I vs Trials')
    plt.show()
#     return y
    
def I_trials2(N,trials):
    print('N:',N,' Trials:',trials)
    y = []
    x = list(range(1, 101))
    for _ in range(1,trials+1):
        dx = 1/N
        I1 = sum([3*(random.random()**2)*dx for __ in range(N)])
        I2 = 0
        for __ in range(N):
            yy = random.random()
            i = sum([(random.random()**2)*yy*dx for ___ in range(N)])
            I2 += i * dx
        y.append(I2)
    print('Integral:',y[-1])
    print('Standard deviation:',np.std(y))
    plt.plot(y,label="Simulated")
    plt.plot([1/6 for i in range(trials)],label="Actual")
    plt.xlabel("Trials")
    plt.ylabel("I")
    plt.legend()
    plt.title('I vs Trials')
    plt.show()
```

```python
def deviate(N,T):
    y = []
    x = list(range(1, N+1))
    for n in range(1, N+1):
        t = np.zeros(T)
        for j in range(T):
            dx = 1/n
            I = sum([3*(random.random()**2)*dx for __ in range(n)])
            t[j] = I
        y.append(np.std(t))
    plt.title('Standard Deviation vs N')
    plt.plot(y, label='Std. Dev')
    plt.xlabel("N")
    plt.ylabel("Standard Deviation")
    plt.legend()
    plt.show()

    plt.plot([1/y[i] for i in range(len(x))],label="1/stddev")
    plt.xlabel("N")
    plt.legend()
    plt.title('Standard Deviation vs sqrt(N)')
    plt.show()
    
def deviate2(N,T):
    y = []
    x = list(range(1, N+1))
    for n in range(1, N+1):
        t = np.zeros(T)
        for j in range(T):
            dx = 1/n
            I = 0
            for __ in range(n):
                yy = random.random()
                i = np.sum([(random.random()**2)*yy*dx for ___ in range(n)])
                I += i * dx
            t[j] = I
        y.append(np.std(t))
    plt.title('Standard Deviation vs N')
    plt.plot(y, label="Std. Dev")
    plt.xlabel("N")
    plt.ylabel("Standard Deviation")
    plt.legend()
    plt.show()

    plt.plot([1/i for i in y],label="1/stddev")
    plt.xlabel("N")
    plt.legend()
    plt.title('Standard Deviation vs sqrt(N)')
    plt.show()
```

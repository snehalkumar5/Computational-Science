```
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

## Probability of 2 drunks meeting after n steps


```
n = 100
y = []
x = np.array(range(n))
for i in x:
  tmp = comb(2*i,i)
  tmp1 = (1/2)**(2*i)
  y.append(tmp1*tmp)
plt.plot(x,y,label="Analytical")

ck = []
simcnt=1000
meet = [0 for t in range(n)]
for i in x:
  for j in range(0,simcnt):
    drunk1=0
    drunk2=0
    for k in range(0,i):
      walk1 = np.random.rand()
      walk2 = np.random.rand()
      drunk1 += 1 if walk1 < 0.5 else -1
      drunk2 += 1 if walk2 < 0.5 else -1
    if drunk1==drunk2:
      meet[i]+=1
  ck.append(meet[i]/simcnt)
plt.plot(x,ck,label="Computational")
plt.legend(loc='best')
plt.xlabel("N")
plt.ylabel("Probability")
plt.title("Probability of 2 drunks meeting after n steps")
plt.show()
```

## Probability of drunk returning to origin after n steps

```
n = 100
y = []
x = np.array(range(n))
for i in x:
  tmp = comb(2*i,i)
  tmp1 = (1/2)**(2*i)
  ret = 0 if i%2==1 else tmp*tmp1
  y.append(ret)
plt.plot(x,y,label="Analytical")
simcnt=1000
single=[]
sing=[]
start=0
back=[0 for t in range(n)]
for i in x:
  for j in range(1,simcnt+1):
    step=0
    for k in range(0,i):
      step += 1 if np.random.rand() < 0.5 else -1
    if step == start:
      back[i] += 1
  single.append(back[i]/simcnt)
plt.plot(x,single,label="Computational")
plt.legend(loc="best")
plt.xlabel("N")
plt.ylabel("Probability")
plt.title("Probability of drunk returning to origin after n steps")
plt.show()
```

## Mean displacement of drunk returning after n steps


```
n = 100
x = np.array(range(n))
y = [0 for i in x]
plt.plot(x,y,label="Analytical")
sing=[]
start=0
simcnt=1000
disp=[0 for t in range(n)]
for i in x:
  for j in range(1,simcnt+1):
    step=0
    for k in range(0,i):
      step += 1 if np.random.rand() < 0.5 else -1
    disp[i]+=step-start
  sing.append(disp[i]/simcnt)
plt.plot(x,sing,label="Computational")
plt.legend(loc="best")
plt.ylim([-10,10])
plt.xlabel("N")
plt.ylabel("Mean displacement")
plt.title("Mean displacement of drunk returning after n steps")
plt.show()
```

## Mean squared displacement of drunk returning after n steps


```
n = 100
x = np.array(range(n))
y = [s for s in x]
single=[]
start=0
simcnt=1000
disp=[0 for t in range(n)]
for i in x:
  for j in range(1,simcnt+1):
    step=0
    for k in range(0,i):
      step += 1 if np.random.rand() >= 0.5 else -1
    disp[i]+=(step-start)**2
  single.append(disp[i]/simcnt)
plt.plot(x,y,label="Analytic")
plt.plot(x,single,label="Computational")
plt.legend(loc="best")
plt.xlabel("N")
plt.ylabel("Mean squared displacement")
plt.title("Mean squared displacement of drunk returning after n steps")
plt.show()
```

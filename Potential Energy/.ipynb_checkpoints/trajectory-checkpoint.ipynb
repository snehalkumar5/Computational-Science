# Harmonic Oscillator


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = 0.0001
k = 1
m = 1
sims = 100000
```

```python
for h in range(0,11,2):
    cur_p = np.sqrt(2*m*h)
    cur_x = 0
    tim = 0
    P,X,T = [],[],[]
    
    for i in range(sims):
        T.append(i)
        P.append(cur_p)
        X.append(cur_x)
        tim += dt
        delta_p = -1*k*cur_x*dt
        delta_x = cur_p*dt/m
        cur_p += delta_p
        cur_x += delta_x
    
    plt.plot(X, P, label="H:"+str(h))
    
plt.xlabel('X')
plt.ylabel('P')
plt.legend()
plt.title('Time evolution of Momentum vs Position')
plt.show()
```

```python
for h in range(0,10):
    cur_p = np.sqrt(2*m*h)
    cur_x = 0
    tim = 0
    XX,T = [],[]
    
    for i in range(sims):
        T.append(i)
        XX.append(cur_x**2)
        tim += dt
        delta_p = -1*k*cur_x*dt
        delta_x = cur_p*dt/m
        cur_p += delta_p
        cur_x += delta_x
    
    plt.plot(T, XX, label='H:'+str(h))

plt.title('Mean Squared Displacement vs Time')
plt.xlabel('Time')
plt.ylabel('MSD')
plt.legend()
plt.show()
```
